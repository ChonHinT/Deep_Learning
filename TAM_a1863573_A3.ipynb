# 2. Spliting Data

```python
total_length = len(df)
length_train = round(total_length * 0.7)
length_validation = total_length - length_train
```

```python
df = df.drop("Date", axis=1)
training_data = df.iloc[:length_train, :].reset_index()
validation_data = df.iloc[length_train:, :].reset_index()
```

# 3. Data Pre-Processing

```python
# Scale
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
transformed_training_data = scaler.fit_transform(training_data[['Volume', 'Stock Splits']])
transformed_validation_data = scaler.transform(validation_data[['Volume', 'Stock Splits']])
```

```python
transformed_training_data = pd.DataFrame(transformed_training_data, columns=['Volume', 'Stock Splits']).join(training_data[['Open', 'High', 'Low', 'Close']])
transformed_validation_data = pd.DataFrame(transformed_validation_data, columns=['Volume', 'Stock Splits']).join(validation_data[['Open', 'High', 'Low', 'Close']])
```

```python
transformed_validation_data
```

```python
for column in ['Open', 'High', 'Low', 'Close']:
    transformed_training_data[column] = np.log(getattr(transformed_training_data, column))
    transformed_training_data[column] = transformed_training_data[column].diff() * 10
    transformed_validation_data[column] = np.log(getattr(transformed_validation_data, column))
    transformed_validation_data[column] = transformed_validation_data[column].diff() * 10
```

```python
figure(figsize=(10,4),dpi=100)
plt.plot(transformed_training_data.High)
plt.show()
```

```python
transformed_training_data = transformed_training_data.iloc[1: , :]
transformed_validation_data = transformed_validation_data.iloc[1: , :]
```

```python
transformed_training_data
```

```python
transformed_training_data = transformed_training_data.to_numpy()
transformed_validation_data = transformed_validation_data.to_numpy()
```

```python
transformed_training_data.shape
```

```python
# Sliding window -> generate training set X, y (target on close price)
X_train, X_validation = [], []
y_train, y_validation = [], []

time_step = 50

for time in range(time_step, length_train - 1):
    X_train.append(transformed_training_data[time-time_step:time,:])
    y_train.append(transformed_training_data[time,5])

for time in range(time_step, length_validation - 1):
    X_validation.append(transformed_validation_data[time-time_step:time,:])
    y_validation.append(transformed_validation_data[time,5])

X_train, X_validation = np.array(X_train), np.array(X_validation)
y_train, y_validation = np.array(y_train), np.array(y_validation)
```

# 4. Build Simple RNN Model

```python
from keras.models import Sequential
from keras.layers import Dense, SimpleRNN, LSTM, Dropout, GRU, LayerNormalization
from keras.callbacks import EarlyStopping
from keras.src.layers.core.activation import Activation
import tensorflow as tf
```

```python
simple_model = Sequential()

#simple_model.add(SimpleRNN(128, input_shape=(50, 1), return_sequences=True))
simple_model.add(SimpleRNN(128, input_shape=(50, 6), return_sequences=True))
simple_model.add(LayerNormalization())
simple_model.add(Dropout(0.1))
simple_model.add(Activation('tanh'))

simple_model.add(SimpleRNN(128))
simple_model.add(LayerNormalization())
simple_model.add(Dropout(0.1))
simple_model.add(Activation('tanh'))

#simple_model.add(SimpleRNN(128))
#simple_model.add(Activation('tanh'))

#simple_model.add(SimpleRNN(128))
#simple_model.add(Activation('tanh'))

simple_model.add(Dense(1))

simple_model.summary()
```

```python
optimizer = tf.keras.optimizers.RMSprop(learning_rate=0.0001, clipvalue=1.0)
```

```python
simple_model.compile(loss='mean_squared_error', optimizer=optimizer)
```

```python
early_stopping = EarlyStopping(monitor='loss', patience=3, restore_best_weights=True)
```

```python
simple_history = simple_model.fit(X_train, y_train, validation_data=(X_validation, y_validation), epochs = 30, batch_size = 32, callbacks=[early_stopping])
```

# 6. Hyperparameter Tuning

## Number of Units

```python
best_units = None
best_rmse = float('inf')

units_to_try = [64, 128, 256]

for units in units_to_try:
    simple_model = Sequential()
    simple_model.add(SimpleRNN(units, input_shape=(50, 6), return_sequences=True))
    simple_model.add(LayerNormalization())
    simple_model.add(Dropout(0.1))
    simple_model.add(Activation('tanh'))
    simple_model.add(SimpleRNN(units))
    simple_model.add(LayerNormalization())
    simple_model.add(Dropout(0.1))
    simple_model.add(Activation('tanh'))
    simple_model.add(Dense(1))

    optimizer = tf.keras.optimizers.RMSprop(learning_rate=0.0001, clipvalue=1.0)
    simple_model.compile(loss='mean_squared_error', optimizer=optimizer)
    early_stopping = EarlyStopping(monitor='loss', patience=3, restore_best_weights=True)
    simple_history = simple_model.fit(X_train, y_train, validation_data=(X_validation, y_validation), epochs=30, batch_size=32, callbacks=[early_stopping])

    y_pred = simple_model.predict(X_validation)
    y_pred = np.append(np.repeat(np.nan, 51), y_pred)

    prediction_df['adjustment_factor'] = y_pred
    for i in range(51, len(prediction_df)):
        prediction_df.loc[i, 'Predicted Close'] = prediction_df.loc[i - 1, 'Actual Close'] * math.exp(prediction_df.loc[i, 'adjustment_factor'] / 10)

    df_cleaned = prediction_df.dropna(subset=['adjustment_factor', 'Predicted Close'])

    actual_close = df_cleaned['Actual Close']
    predicted_close = df_cleaned['Predicted Close']
    mse = mean_squared_error(actual_close, predicted_close)
    rmse = round(np.sqrt(mse), 4)

    print()
    print(f'Root Mean Squared Error for {units} units: {rmse}')
    print()

    # Update best_units and best_rmse if the current configuration is better
    if rmse < best_rmse:
        best_units = units
        best_rmse = rmse

print('-' * 50)
print(f'Best number of units: {best_units} with Best RMSE: {best_rmse}')
print('-' * 50)
```

```python
best_layers = None
best_rmse_layers = float('inf')

layers_to_try = [1, 2, 3, 4]

for layers in layers_to_try:
    simple_model = Sequential()
    for _ in range(layers):
        simple_model.add(SimpleRNN(64, input_shape=(50, 6), return_sequences=True))
        simple_model.add(LayerNormalization())
        simple_model.add(Dropout(0.1))
        simple_model.add(Activation('tanh'))

    simple_model.add(SimpleRNN(64))
    simple_model.add(LayerNormalization())
    simple_model.add(Dropout(0.1))
    simple_model.add(Activation('tanh'))
    simple_model.add(Dense(1))

    optimizer = tf.keras.optimizers.RMSprop(learning_rate=0.0001, clipvalue=1.0)
    simple_model.compile(loss='mean_squared_error', optimizer=optimizer)

    early_stopping = EarlyStopping(monitor='loss', patience=3, restore_best_weights=True)
    simple_history = simple_model.fit(X_train, y_train, validation_data=(X_validation, y_validation), epochs=30, batch_size=32, callbacks=[early_stopping])

    y_pred = simple_model.predict(X_validation)
    y_pred = np.append(np.repeat(np.nan, 51), y_pred)

    prediction_df['adjustment_factor'] = y_pred
    for i in range(51, len(prediction_df)):
        prediction_df.loc[i, 'Predicted Close'] = prediction_df.loc[i - 1, 'Actual Close'] * math.exp(prediction_df.loc[i, 'adjustment_factor'] / 10)

    df_cleaned = prediction_df.dropna(subset=['adjustment_factor', 'Predicted Close'])

    actual_close = df_cleaned['Actual Close']
    predicted_close = df_cleaned['Predicted Close']
    mse = mean_squared_error(actual_close, predicted_close)
    rmse = round(np.sqrt(mse), 4)

    print()
    print(f'Root Mean Squared Error for {layers} layers: {rmse}')
    print()

    # Update best_layers and best_rmse_layers if the current configuration is better
    if rmse < best_rmse_layers:
        best_layers = layers
        best_rmse_layers = rmse

print('-' * 50)
print(f'Best number of units: {best_units} with Best number of layers: {best_layers} and Best RMSE: {best_rmse_layers}')
print('-' * 50)
```

## Optimizers & its default learning rate

```python
best_units = 64
best_layers = 4
best_dropout = 0.2
best_optimizer = None
best_rmse_optimizer = float('inf')

optimizers_to_try = {'Adam': 0.001, 'SGD': 0.01, 'RMSprop': 0.001}

for optimizer_name, learning_rate in optimizers_to_try.items():
    simple_model = Sequential()

    for _ in range(best_layers):
        simple_model.add(SimpleRNN(best_units, input_shape=(50, 6), return_sequences=True))
        simple_model.add(LayerNormalization())
        simple_model.add(Dropout(best_dropout))
        simple_model.add(Activation('tanh'))

    simple_model.add(SimpleRNN(best_units))
    simple_model.add(LayerNormalization())
    simple_model.add(Dropout(best_dropout))
    simple_model.add(Activation('tanh'))

    # Add the final Dense layer
    simple_model.add(Dense(1))

    optimizer = getattr(tf.keras.optimizers, optimizer_name)(learning_rate=learning_rate, clipvalue=1.0)
    simple_model.compile(loss='mean_squared_error', optimizer=optimizer)

    early_stopping = EarlyStopping(monitor='loss', patience=3, restore_best_weights=True)
    simple_history = simple_model.fit(X_train, y_train, validation_data=(X_validation, y_validation), epochs=30, batch_size=32, callbacks=[early_stopping])

    y_pred = simple_model.predict(X_validation)
    y_pred = np.append(np.repeat(np.nan, 51), y_pred)

    prediction_df['adjustment_factor'] = y_pred
    for i in range(51, len(prediction_df)):
        prediction_df.loc[i, 'Predicted Close'] = prediction_df.loc[i - 1, 'Actual Close'] * math.exp(prediction_df.loc[i, 'adjustment_factor'] / 10)

    df_cleaned = prediction_df.dropna(subset=['adjustment_factor', 'Predicted Close'])

    actual_close = df_cleaned['Actual Close']
    predicted_close = df_cleaned['Predicted Close']
    mse = mean_squared_error(actual_close, predicted_close)
    rmse = round(np.sqrt(mse), 4)

    print()
    print(f'Root Mean Squared Error for {optimizer_name.capitalize()} optimizer with Learning Rate {learning_rate}: {rmse}')
    print()

    # Update best_optimizer and best_rmse_optimizer if the current configuration is better
    if rmse < best_rmse_optimizer:
        best_optimizer = optimizer_name
        best_rmse_optimizer = rmse

print('-' * 50)
print(f'Best number of units: {best_units} with Best number of layers: {best_layers}, Best Dropout Rate: {best_dropout}, Best Optimizer: {best_optimizer.capitalize()} with Default Learning Rate and Best RMSE: {best_rmse_optimizer}')
print('-' * 50)
```

## Gradient clipping

```python
best_units = 64
best_layers = 4
best_dropout = 0.2
best_optimizer = 'SGD'
best_learning_rate = 0.01

clip_value_range = [1.0, 2.0]

best_clip_value = None
best_rmse_clip_value = float('inf')

for clip_value in clip_value_range:
    simple_model = Sequential()

    for _ in range(best_layers):
        simple_model.add(SimpleRNN(best_units, input_shape=(50, 6), return_sequences=True))
        simple_model.add(LayerNormalization())
        simple_model.add(Dropout(best_dropout))
        simple_model.add(Activation('tanh'))

    simple_model.add(SimpleRNN(best_units))
    simple_model.add(LayerNormalization())
    simple_model.add(Dropout(best_dropout))
    simple_model.add(Activation('tanh'))

    simple_model.add(Dense(1))

    optimizer = tf.keras.optimizers.SGD(learning_rate=best_learning_rate, clipvalue=clip_value)
    simple_model.compile(loss='mean_squared_error', optimizer=optimizer)

    early_stopping = EarlyStopping(monitor='loss', patience=3, restore_best_weights=True)
    simple_history = simple_model.fit(X_train, y_train, validation_data=(X_validation, y_validation), epochs=30, batch_size=32, callbacks=[early_stopping])

    y_pred = simple_model.predict(X_validation)
    y_pred = np.append(np.repeat(np.nan, 51), y_pred)

    prediction_df['adjustment_factor'] = y_pred
    for i in range(51, len(prediction_df)):
        prediction_df.loc[i, 'Predicted Close'] = prediction_df.loc[i - 1, 'Actual Close'] * math.exp(prediction_df.loc[i, 'adjustment_factor'] / 10)

    df_cleaned = prediction_df.dropna(subset=['adjustment_factor', 'Predicted Close'])

    actual_close = df_cleaned['Actual Close']
    predicted_close = df_cleaned['Predicted Close']
    mse = mean_squared_error(actual_close, predicted_close)
    rmse = round(np.sqrt(mse), 4)

    print('-' * 50)
    print(f'Root Mean Squared Error for SGD optimizer with Learning Rate {best_learning_rate}, clipvalue={clip_value}: {rmse}')
    print('-' * 50)

    if rmse < best_rmse_clip_value:
        best_clip_value = clip_value
        best_rmse_clip_value = rmse

print('-' * 50)
print(f'Best number of units: {best_units}, Best number of layers: {best_layers}, Best Dropout Rate: {best_dropout}, Best Optimizer: {best_optimizer} with Learning Rate {best_learning_rate}, Best Clip Value: {best_clip_value}, and Best RMSE: {best_rmse_clip_value}')
print('-' * 50)

```
